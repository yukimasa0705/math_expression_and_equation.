# 一次不等式とは

一次不等式とは、**変数が1次式**で表される不等式のことを言います。  
例えば、

- $2x + 3 > 7$
- $5x - 1 \leq 14$
- $-3x + 2 \geq -4$

などが一次不等式です。

# 解き方の基本

一次不等式は、**方程式とほぼ同じ感覚で解けます**。  
ただし、注意点がひとつあります。

> 不等式の両辺を**負の数で掛けたり割ったりしたとき**、不等号の向きが逆になります！

このポイントをしっかり押さえておきましょう。

# 基本の手順

1. 文字を含む項と定数項を整理する  
2. 必要なら両辺を同じ数で割る or 掛ける  
3. 答えを書くときは、不等号に注意！

# 例題

例1） $2x + 3 > 7$ を解こう。

まず、定数項を移項する。
\[
2x > 7 - 3
\]
\[
2x > 4
\]

次に、両辺を2で割る。
\[
x > 2
\]

よって、答えは $x > 2$ です。

# ポイントまとめ

- 負の数で掛けたり割ったりしたら、不等号をひっくり返す
- 方程式と同じ感覚でOK、ただし不等号を忘れずに！


# 連立一次不等式とグラフによる解き方

連立一次不等式とは、**複数の一次不等式を同時に満たすような解を求める**問題です。  
グラフを使って、解の範囲を視覚的に確認することもできます。

# 例題

次の連立一次不等式を解き、グラフに表しましょう。

$$
\begin{cases}
2x - 1 > 3 \\
- x + 4 \geq 1
\end{cases}
$$

## Step 1. 個別に解く

まずそれぞれの不等式を解きます。

### 1つ目

$$
2x - 1 > 3
$$

両辺に1を足す。

$$
2x > 4
$$

両辺を2で割る。

$$
x > 2
$$

---

### 2つ目

$$
- x + 4 \geq 1
$$

両辺から4を引く。

$$
- x \geq -3
$$

両辺を**-1倍**する（不等号の向きが変わる！）。

$$
x \leq 3
$$

---

## Step 2. 共通する範囲を探す

つまり、

$$
2 < x \leq 3
$$

が解となります。

## Step 3. グラフに表す

以下のコードで、グラフを描画してみましょう。

```python
import matplotlib.pyplot as plt
import numpy as np

# xの範囲を設定
x = np.linspace(0, 5, 400)

# 不等式のグラフを描く
y1 = 2*x - 1 - 3  # 2x - 1 > 3
y2 = -x + 4 - 1   # -x + 4 ≥ 1

# グラフをプロット
plt.figure(figsize=(6,6))
plt.plot(x, y1, label='$2x - 1 > 3$', color='blue')
plt.plot(x, y2, label='$- x + 4 ≥ 1$', color='green')

# 解の範囲を視覚化
plt.fill_between(x, 0, y1, where=(x > 2), color='blue', alpha=0.3)
plt.fill_between(x, 0, y2, where=(x <= 3), color='green', alpha=0.3)

# 解の交差部分（2 < x <= 3）を塗りつぶす
plt.fill_between(x, 0, y1, where=(x > 2) & (x <= 3), color='purple', alpha=0.6, label='$2 < x ≤ 3$')

# グラフの設定
plt.axhline(0, color='black',linewidth=1)
plt.axvline(0, color='black',linewidth=1)
plt.xlim(0, 5)
plt.ylim(-5, 5)
plt.xlabel('x')
plt.ylabel('y')
plt.title('連立一次不等式のグラフ')
plt.legend()

plt.show()
```

# 二元一次連立不等式とグラフによる解き方

二元一次連立不等式とは、**2つの変数（x, y）を含む一次式の不等式**を同時に満たす解を求める問題です。  
この場合、2つの直線で囲まれる範囲が解になります。

# 例題

次の二元一次連立不等式を解き、グラフに表しましょう。

$$
\begin{cases}
x + y \geq 4 \\
2x - y \leq 3
\end{cases}
$$

## Step 1. 個別に解く

まず、それぞれの不等式を解きます。

### 1つ目

$$
x + y \geq 4
$$

この式は、**直線 $x + y = 4$** より上の領域です。

---

### 2つ目

$$
2x - y \leq 3
$$

この式は、**直線 $2x - y = 3$** より下の領域です。

## Step 2. グラフに表す

以下のコードで、この二元一次連立不等式をグラフに表しましょう。


```python
import numpy as np
import matplotlib.pyplot as plt

# x, yの範囲を設定
x = np.linspace(-1, 5, 400)
y1 = 4 - x  # x + y >= 4 -> y >= 4 - x
y2 = 2*x - 3  # 2x - y <= 3 -> y >= 2x - 3

# グラフの描画
plt.figure(figsize=(6,6))

# 不等式の領域を描画
plt.fill_between(x, y1, 5, where=(y1 >= 0), color='blue', alpha=0.3, label='$x + y ≥ 4$')
plt.fill_between(x, -5, y2, where=(y2 <= 5), color='green', alpha=0.3, label='$2x - y ≤ 3$')

# 共通する範囲 (交差部分)
plt.fill_between(x, np.maximum(y1, y2), 5, where=(y1 >= y2), color='purple', alpha=0.6, label='共通する範囲')

# グラフの設定
plt.axhline(0, color='black',linewidth=1)
plt.axvline(0, color='black',linewidth=1)
plt.xlim(-1, 5)
plt.ylim(-1, 5)
plt.xlabel('x')
plt.ylabel('y')
plt.title('二元一次連立不等式のグラフ')

# 凡例を表示
plt.legend()

plt.show()
```

# 解説

青い領域は $x + y \geq 4$ を表しており、直線 $y = 4 - x$ より上の範囲を示します。

緑の領域は $2x - y \leq 3$ を表しており、直線 $y = 2x - 3$ より上の範囲を示します。

紫の領域が、二元一次連立不等式の両方を満たす解の範囲です。

# グラフから解を導き出す

グラフでは、**交差部分（紫の領域）**がこの二元一次連立不等式を同時に満たす解になります。  
具体的には、この領域が $y \geq 4 - x$ と $y \geq 2x - 3$ の両方を満たす範囲です。

# 交差部分を式で導く

グラフからこの交差部分がどこかを見ると、両方の不等式を満たす範囲は次のように求められます。

1. 最初の不等式 $x + y \geq 4$ の式は、直線 $y = 4 - x$ より上の領域です。
2. 次の不等式 $2x - y \leq 3$ の式は、直線 $y = 2x - 3$ より上の領域です。

そのため、両方の不等式を満たす解は、交差部分に位置する点、すなわち $y$ が $4 - x$ 以上、かつ $2x - 3$ 以下である範囲です。

# 解の範囲

$x$ が $-1$ から $5$ の間で、その間にある $y$ の範囲が解となります。

このように、グラフから解の範囲を視覚的に確認し、式としても導き出すことができます。

# ポイントまとめ

- 不等式をそれぞれグラフにして、領域を描画する
- 両方の不等式を満たす範囲は、グラフの交差部分である
- 交差部分の範囲が解の範囲に対応する
